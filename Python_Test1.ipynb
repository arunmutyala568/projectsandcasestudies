```python
def myfun(myarray):
    
    for i in myarray:
        
        if type(i)==str:
            
            try:
                
                raise ValueError("Occured at",i)
                
            except ValueError as t:
                
                print("oops")
                print(type(t))
                print(t.args)
                
        elif i==0:
            
            try:
                
                raise ZeroDivisionError("occured at",i)
                
            except ZeroDivisionError as t:
                
                print("oops")
                
                print(type(t))
                
                print(t.args)
            
        elif type(i)==int:
            print("reciprocal")
            print(1/i)
                
a=[0,"arun",2]

myfun(a)
```
