```python
transactionsdataframe=pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')

transactionsdataframe
```

```python
mysql("""select 

         transactionsdataframe.TransactionID,transactionsdataframe.TransactionDate,transactionsdataframe.UserID,

         transactionsdataframe.ProductID,transactionsdataframe.Quantity,usersdataframe.User,usersdataframe.Gender,usersdataframe.Registered,
         
         usersdataframe.Cancelled
         
         from transactionsdataframe 
         
         LEFT JOIN usersdataframe on transactionsdataframe.UserID=usersdataframe.UserID""")
```

```python
mysql("""SELECT transactionsdataframe.TransactionID,transactionsdataframe.TransactionDate,transactionsdataframe.ProductID,transactionsdataframe.UserID
        ,transactionsdataframe.Quantity FROM transactionsdataframe
          WHERE NOT EXISTS
           (SELECT *
            FROM usersdataframe
            WHERE usersdataframe.UserID= transactionsdataframe.UserID
            )""")
```

```python
mysql("""SELECT   transactionsdataframe.TransactionID,transactionsdataframe.TransactionDate,transactionsdataframe.UserID,

         transactionsdataframe.ProductID,transactionsdataframe.Quantity,usersdataframe.User,usersdataframe.Gender,usersdataframe.Registered,
         
         usersdataframe.Cancelled
         FROM transactionsdataframe
         INNER JOIN usersdataframe ON transactionsdataframe.UserID = usersdataframe.UserID ORDER BY User
         ;
""")
```

```python
mysql("""SELECT *
          FROM transactionsdataframe
          FULL OUTER JOIN usersdataframe
          ON transactionsdataframe.UserID= usersdataframe.UserID
          ORDER BY usersdataframe.User ;""")
```
